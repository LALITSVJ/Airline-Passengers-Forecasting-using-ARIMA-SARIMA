### Augmented Dickeyâ€“Fuller (ADF) Test
We are checking stationarity of both generated curve using Augmented Dickey-Fuller-Test.

```python
# Importing necessary dependancies
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
# Read the both excel files
df_seasonal = pd.read_csv('airline_passengers_with_Seasonality.csv')
df_without_seasonal = pd.read_csv('airline_passengers_Without_Seasonality.csv')
```

```python
# ADF Test score for df_seasonal
result_seasonal = adfuller(df_seasonal['Passengers'])
print('ADF Test results are for seasonal airline passengers data: ')
print(result_seasonal)

print('\nUseful results for us: ')
print('ADF Statistic: %f' % result_seasonal[0])
print('p-value: %f' % result_seasonal[1])
print('Critical Values:')
for key, value in result_seasonal[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# ADF Test score for df_non-seasonal
result_non_seasonal = adfuller(df_without_seasonal['Passengers'])
print('ADF Test results are for non-seasonal airline passengers data: ')
print(result_non_seasonal)

print('\nUseful results for us: ')
print('ADF Statistic: %f' % result_non_seasonal[0])
print('p-value: %f' % result_non_seasonal[1])
print('Critical Values:')
for key, value in result_non_seasonal[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# As we can observe, p-value is less than 0.005 means the data generated by removing seasonality is stationary now and can be provided to models for forecasting. 

# Thus ADF Test analysis has been completed.
```
